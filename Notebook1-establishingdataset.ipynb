# Describing Data Features

```python
print(f"The number of rows and columns, respectively, in the dataset is {AI_trends.shape}.")
```

```python
COLS = pd.DataFrame({
    "Column Name": AI_trends.columns,
    "Data Type": [AI_trends[col].dtype for col in AI_trends.columns],
    "First 3 Possible Values": [AI_trends[col].unique()[:3] for col in AI_trends.columns],
    "# of Unique Values": [AI_trends[col].nunique() for col in AI_trends.columns]
})
COLS
```

Exploring the Numeric Columns:

```python
numeric_cols = COLS[COLS['Data Type'] != 'object']['Column Name']

num_COLS = pd.DataFrame({
    "Column Name": numeric_cols,
    "Minimum Value": [AI_trends[col].min() for col in numeric_cols],
    "Maximum Value": [AI_trends[col].max() for col in numeric_cols],
    "Mean Value": [AI_trends[col].mean() for col in numeric_cols],
    "Median Value": [AI_trends[col].median() for col in numeric_cols]
})

num_COLS
```
